# CREATING DUMMY VARIEABLES 

```python
data=pd.get_dummies(data,dummy_na=False,columns=['Engine Fuel Type','Year','Engine Cylinders', 'Transmission Type','Driven_Wheels','Vehicle Size','Vehicle Style'],drop_first=True)
```

```python
data.head()
```

# MODEL TRAINING

```python
#split into test and train 

from sklearn.model_selection import train_test_split
X=data.drop(['MSRP'],axis=1)
y=data['MSRP']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=13)
```

```python
from sklearn.ensemble import RandomForestRegressor
lr=RandomForestRegressor(n_estimators=500,random_state=13)
model=lr.fit(X_train,y_train)
prediction=model.predict(X_test)
```

# Accuracy CHECK

```python
errors=abs(prediction-y_test)
print('Mean Absolute Error:', round(np.mean(errors), 2))
```

```python
MAPE=100*(errors/prediction)
Accuracy=100-np.mean(MAPE)
print('Accuracy:', round(Accuracy,2),'%')
```
